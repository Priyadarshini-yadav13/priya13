```python
#creating a dataframe
df1=pd.DataFrame({
    'key': ['K0','K1','K2','K3'],
    'key1': ['K0','K1','K2','K3'],
    'Name':['Pankaj','Sai','Niraj','Vijay'],
    'Age': [27,28,29,30]
})
df2=pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'key1': ['K0','K1','K2','K3'],
    'Address':['Nagpur','Kanpur','Satara','Pune'],
    'Qualification':['MA','MCOM','MCA','BSc'],
})
print (df1, '\n\n',df2)
```

```python
#merge with multiple keys
res=pd.merge(df1,df2,on=['key','key1'])
res
```

```python
#creating a dataframe
df1=pd.DataFrame({
    'key': ['K0','K1','K2','K3'],
    'key1': ['K0','K1','K0','K1'],
    'Name':['Pankaj','Sai','Niraj','Vijay'],
    'Age': [27,28,29,30]
})
df2=pd.DataFrame({
    'key':['K0','K1','K2','K3'],
    'key1': ['K0','K0','K0','K0'],
    'Address':['Nagpur','Kanpur','Satara','Pune'],
    'Qualification':['MA','MCOM','MCA','BSc'],
})
print (df1, '\n\n',df2)
```

```python
#merge with composite keys
res=pd.merge(df1,df2,how='outer',on=['key','key1'])
res
```

```python
#in how=outer all key combination comes.the value which is not available comes nan..
```

```python
#merge with composite keys
res=pd.merge(df1,df2,how='inner',on=['key','key1'])
res
```

```python
#with inner we get only common one,by default value is inner.
```

```python
#merge with composite keys
res=pd.merge(df1,df2,how='left',on=['key','key1'])
res
```

```python
#we are going to take key,key1 from df1 as how=left and
 we get null value if key,key1 doesnot match with df
```

```python
#merge with composite keys
res=pd.merge(df1,df2,how='right',on=['key','key1'])
res
```

```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Name':['Jane','John','Ashley','Mike','Emily','Jack'],
    'Ctg': ['A', 'A', 'C' ,'B','B','C'],
     'Val': np.random.random(6).round(2),
     'Val2': np.random.randint(1, 10, size=6)
})
df.head()
```

```python
#filtering according to logical operator
df[df['Val'] > 0.5]
```

```python
#logical operator works on string values
df[df['Name']>'Jane']
```

```python
#alphabettically all the name after jane will retuned.
```

```python
#Adding the multiple conditioning while filtering
```

```python
#Adding the multiple conditioning while filtering(OR operator),,,,|line is used
df[(df['Val'] > 0.5) | (df['Val2']==5)]
```

```python
#or function gives eithor one is true or another one is true or both are true.at least one condition should be true.
```

```python
#Adding the multiple conditioning while filtering(and function)
df[(df['Val'] > 0.5) & (df['Val2']==9)]
```

```python
#in and operation both the condition should be true.
```

```python
#isin()operator
names=['John','Mike']
df[df['Name'].isin(names)]
```

```python
#string accessor
df[df['Name'].str.startswith('J')]
```

```python
#creating dataframe
import pandas as pd
df=pd.DataFrame({
'Product':['carrot','broccoli','banana','banana',
          'Beans','orange','carrot','broccoli'],
'Category':['Vegetable','Vegetable','Fruit','Fruit',
          'Vegetable','Fruit','Vegetable','Vegetable'],
'Quantity':[8,5,3,4,5,9,11,8],
'Amount'  :[270,240,620,384,640,610,620,190]})
df.head()
```

```python
#Pivot table for sum aggregation
df.pivot_table(index=['Product'],
               values=['Amount'],
               aggfunc='sum')

```

```python
#Pivot table for sum aggregation
pivot=df.pivot_table(index=['Product','Category'],
               values=['Amount'],
               aggfunc='sum')
print(pivot)
```

```python
#Pivot table for sum aggregation
Pivot=df.pivot_table(index=['Product','Category'],
               values=['Amount','Quantity'],
               aggfunc='sum')
print(Pivot)
```

```python
#Pivot table for sum aggregation
Pivot=df.pivot_table(index=['Product','Category'],
               values=['Amount','Quantity'],
               aggfunc=['sum','median'])
print(Pivot)
```
