```python
#catplot-categorical scatterplot
#Getting the data
exercise=sns.load_dataset('exercise',index_col=0)
exercise.head()
```

```python
sns.get_dataset_names()
```

```python
#catplot(continued...)
sns.catplot(x='time',y='pulse',data=exercise,hue='kind')
```

```python
#distplot()
plt.figure(figsize=(5,3))
sns.distplot(exercise['pulse'],kde=False,color='red')
```

```python
#distplot()
plt.figure(figsize=(5,3))
sns.distplot(exercise['pulse'],kde=True,color='Green')
```

```python
#performance comparison
#importing time
import time
#size of array and lists
size=1000000
#declaring two lists
List1=range(size)
List2=range(size)
#declaring two numpy arrays
array1=np.arange(size)
array2=np.arange(size)
#capturing the time before the operation
initialTime=time.time()
#multiplying the elements
res=[(a*b) for a,b in zip(List1,List2)]
#capturing the time after operation
print('Time taken by Lists to perform multiplication:',(time.time()-initialTime),'ms')
#capturing the time before the operation
initialTime=time.time()
#multipling the arrays
res=array1 * array2
print('Time taken by numpy array to perform multiplication:',(time.time()-initialTime),'ms')



```
