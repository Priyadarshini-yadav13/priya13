```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#scatter plot
#creating the data
x=[5,7,8,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,100,86,103,87,94,78,77,85]
plt.figure(figsize=(5,3))
plt.scatter(x,y,color='red')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Scatter Plot')

plt.show()
```

```python
#line plot
#creating the data
year=[1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020]
unemployment_rate=[9.8,12,8,7.2,6.9,7,6.5,6.2,5.5,.3,9.9]
#plotting line plot
plt.figure(figsize=(5,3))
plt.plot(year,unemployment_rate)
plt.xlabel('year')
plt.ylabel('unemployment rate')
plt.title('Unemployment Rate across years')
plt.show()
```

```python
#Bar chart
#creating the data
x=['C','C++','Java','Python','PHP']
y=[22.2,17.6,8.8,8,7.7]
#setting the figure size
plt.figure(figsize=(5,3))
#plotting the bar chart
plt.bar(x,y,color='black',width=0.6)
plt.xlabel('Programming Languages')
plt.ylabel('no of students')
plt.title('Popularity of Programming Languages')
```

```python
#loading an inbuilt dataset
df=sns.load_dataset('tips')
df.head()
```

```python
#Countplot()
#setting the figure size
plt.figure(figsize=(5,3))
#plotting the countplot
sns.countplot(x='day',data=df)
plt.xlabel('Day')
plt.ylabel('Count')
```

```python
#boxplot
plt.figure(figsize=(5,3))
plt.boxplot(df['total_bill'])
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
```

```python
#histogram
plt.figure(figsize=(5,3))
plt.hist(df['total_bill'],bins=30)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
```

```python
#stacked Bar Chart
#creating the data
x=['A','B','C','D']
y1=np.array([10,20,10,30])
y2=np.array([20,2,15,25])
y3=np.array([12,15,19,6])
y4=np.array([10,25,15,20])

#plot bars in stacked manner

plt.bar(x, y1, color ='green')
plt.bar(x, y2, bottom= y1, color ='red')
plt.bar(x, y3, bottom= y1 + y2, color ='blue')
plt.bar(x, y4, bottom= y1 + y2 + y3, color ='yellow')

plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.title('Stacked Bar Chart')
plt.show()
```

```python
#Piechart
#Creating the data
labels={'Python','C++','Ruby','Java'}
sizes=[215,130,245,210]

#plot
plt.figure(figsize=(5,3))
plt.pie(sizes,labels=labels,autopct='%.2f%%',startangle=120)
plt.title('Pie Chart')
plt.show()
```
