```python
#imorting important libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
beml_df=pd.read_csv('BEML.csv')
beml_df.head()
```

```python
#importing the data
glaxo_df=pd.read_csv('GLAXO.csv')
glaxo_df.head()
```

```python
#Filtering the required columes
beml_df=beml_df[['Date','Close']]
glaxo_df=glaxo_df[['Date','Close']]
```

```python
#Getting the datatype
beml_df.dtypes
```

```python
#getting the datatype
glaxo_df.dtypes
```

```python
#Creating a index
beml_df=beml_df.set_index(pd.DatetimeIndex(beml_df['Date']))
beml_df.head()
```

```python
glaxo_df=glaxo_df.set_index(pd.DatetimeIndex(glaxo_df['Date']))
glaxo_df.head()
```

```python
#Data visualisation
plt.figure(figsize=(5,3))
plt.plot(beml_df.index,beml_df['Close'])
plt.xlabel('Year')
plt.ylabel('Closing prize')
plt.title('Closing prize of BEML')
plt.show()
```

```python
#Data visualisation
plt.figure(figsize=(5,3))
plt.plot(glaxo_df.index,glaxo_df['Close'])
plt.xlabel('Year')
plt.ylabel('Closing prize')
plt.title('Closing prize of Glaxo')
plt.show()
```

```python
#Adding a 'gain'column
beml_df['gain']=beml_df.Close.pct_change(periods=1)
beml_df.head()
```

```python
glaxo_df['gain']=glaxo_df.Close.pct_change(periods=1)
glaxo_df.head()
```

```python
#Dropping the first row with NaN value
beml_df.dropna(inplace=True)
glaxo_df.dropna(inplace=True)
```

```python
glaxo_df.head()
```

```python
#plot the gains
plt.figure(figsize=(5,3))
plt.plot(beml_df.index,beml_df['gain'])
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```

```python
#plot the gains
plt.figure(figsize=(5,3))
plt.plot(glaxo_df.index,glaxo_df['gain'])
plt.xlabel('Year')
plt.ylabel('Gain')
plt.title('Gain of BEML')
plt.show()
```

```python
#density plot for gain of glaxo
plt.figure(figsize=(5,3))
sns.displot(glaxo_df['gain'],label='glaxo',kde=True)
plt.xlabel('Gain')
plt.ylabel('Density')
plt.title('Density plot for gain of Glaxo')
plt.show()
```

```python
#density plot for gain of glaxo
plt.figure(figsize=(5,3))
sns.displot(beml_df['gain'],label='beml',kde=True)
plt.xlabel('Gain')
plt.ylabel('Density')
plt.title('Density plot for gain of beml')
plt.show()
```

```python
beml_df['gain'].describe()
```

```python
#getting the stastistical properties
print('Mean :',round(beml_df['gain'].mean(),4))
print('Median :',round(beml_df['gain'].median(),4))
print('Standard Deviation :',round(beml_df['gain'].std(),4))
```

```python
#getting 95% confidence interval for glaxo_df
from scipy import stats
import numpy as np
glaxo_df.ci=stats.norm.interval(0.95,loc=glaxo_df.gain.mean(),scale=glaxo_df.gain.std())
#confidence interval
print('glaxo confidence interval:',np.round(glaxo_df.ci,4))
```

```python
#getting 95% confidence interval for beml_df
beml_df.ci=stats.norm.interval(0.95,loc=beml_df.gain.mean(),scale=beml_df.gain.std())
#confidence interval
print('glaxo confidence interval:',np.round(beml_df.ci,4))
```
